```python
from PIL import Image
import os

# Specify the path to the folder containing the images
folder_path = (r"C:\Users\Hp\Pictures") 

# Get a list of all files in the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]

# Sort the image files in a specific order (e.g., alphabetical order)
image_files.sort()

# Specify the target width and height for all images
target_width, target_height = 200, 200  # Adjust as needed

# Create a list to store the opened and resized images
images = []

# Open, resize, and append each image to the list
for file in image_files:
    image_path = os.path.join(folder_path, file)
    img = Image.open(image_path)
    resized_img = img.resize((target_width, target_height))
    images.append(resized_img)

# Calculate the number of rows needed
num_rows = len(images) // 3 + (len(images) % 3 > 0)  # Ceiling division

# Calculate the size of the result image
result_width = 3 * target_width
result_height = num_rows * target_height

result_image = Image.new('RGB', (result_width, result_height), color='white')

# Paste images into the result image
current_row, current_col = 0, 0
for img in images:
    result_image.paste(img, (current_col * target_width, current_row * target_height))
    current_col += 1
    if current_col == 3:
        current_col = 0
        current_row += 1

# Save or display the result image
result_image.save('result_image.jpg')
result_image.show()
```

```python
import pandas as pd 
import os 
from skimage.transform import resize 
from skimage.io import imread 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import svm 
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
```

```python
Categories=['cats','dogs'] 
flat_data_arr=[] #input array 
target_arr=[] #output array 
datadir='Downloads/dataset/'
#path which contains all the categories of images 
for i in Categories: 
	
	print(f'loading... category : {i}') 
	path=os.path.join(datadir,i) 
	for img in os.listdir(path): 
		img_array=imread(os.path.join(path,img)) 
		img_resized=resize(img_array,(150,150,3)) 
		flat_data_arr.append(img_resized.flatten()) 
		target_arr.append(Categories.index(i)) 
	print(f'loaded category:{i} successfully') 
flat_data=np.array(flat_data_arr) 
target=np.array(target_arr)
```

```python
#dataframe 
df=pd.DataFrame(flat_data) 
df['Target']=target 
df.shape
```

```python
print(flat_data)
```

```python
#input data 
x=df.iloc[:,:-1] 
#output data 
y=df.iloc[:,-1]
```

```python
# Splitting the data into training and testing sets 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10, 
											random_state=77, 
											stratify=y) 
```

```python
# Defining the parameters grid for GridSearchCV 
param_grid={'C':[0.1,1,10,100], 
			'gamma':[0.0001,0.001,0.1,1], 
			'kernel':['rbf','poly']} 

# Creating a support vector classifier 
svc=svm.SVC(probability=True) 

# Creating a model using GridSearchCV with the parameters grid 
model=GridSearchCV(svc,param_grid)
```

```python
# Training the model using the training data 
model.fit(x_train,y_train)
```

```python
# Testing the model using the testing data 
y_pred = model.predict(x_test) 

# Calculating the accuracy of the model 
accuracy = accuracy_score(y_pred, y_test) 

# Print the accuracy of the model 
print(f"The model is {accuracy*100}% accurate")
```

```python
print(classification_report(y_test, y_pred, target_names=['cat', 'dog']))
```
